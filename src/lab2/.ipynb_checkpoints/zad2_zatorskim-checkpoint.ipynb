```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.interpolate import CubicSpline
```

# Ex. 1)


Dla funkcji  $ f(x) = \frac{1}{1 + x^2} $  i $ n = 5, 10, 15 $ wylicz wielomian interpolacyjny $ p_n$ z $ n + 1 $ węzłami równoodległymi w przedziale $ [−5, 5] $.

    a) Narysuj funkcje $f(x)$ oraz $p_n(x)$
    b) Wylicz i narysuj reszty interpolacji $ e(x) = f(x) − p_n(x)$ w 30 równoodległych
    punktach powyższego przedziału.

## Define function and x range

Zapisujemy przedział dla x-ów, wartości n oraz funkcję z treści zadania.

```python
def f(x): return 1/(1 + x**2)

n_cases = [5, 10, 15]
x_range = [-5, 5]
```

## 1 a)

- Mając przedział $ [-5, 5] $, dzielimy go na $ n $ równych części, tworząc $ n + 1 $ węzłów (linie 1-2 funkcji calculate_factors). Następnie dla każdego węzła o współrzędnej x wyliczamy wartość f(x) (używając funkcji zdefiniowanej powyżej).
- Przekazując *wektor współrzędnych x* węzłów do funkcji `vander` z biblioteki `numpy`, otrzymujemy jako wynik **macierz Vandermonde'a**, której użyjemy do obliczenia współczynników naszego wielomianu. 
- Przekazujemy macierz Vandermonde'a (`vander_matrix`) wraz z *wartościami y węzłów* (`y_list` jako wektor) do funkcji `numpy.linalg.solve` i jako wynik otrzymujemy **wektor** zawierający **współczynniki wielomianu interpolacyjnego $ p_n $**.

```python
def calculate_factors(n):
    x_list = np.arange(x_range[0], x_range[1], (x_range[1] - x_range[0]) / n)
    x_list = np.append(x_list, x_range[1])
    y_list = f(x_list)
    
    vander_matrix = np.vander(x_list, increasing=True)
    a_list = np.linalg.solve(vander_matrix, np.transpose(y_list))
    return a_list
```

Mając wektor współczynników, funkcja poniżej zwraca wartość wielomianu w punkcie x.

```python
def calculate_polynomial(a_list, x):
    sum = 0
    for i in range(len(a_list)):
        sum += a_list[i]*(x**i)
    return sum
```

Teraz mając wszystkie potrzebne dane, chcemy na wykresie przedstawić zarówno funkcję $ f(x) $ jak i wielomian interpolacyjny $p_n$. Poniższa funkcja realizuje tę zadanie. Jako parametr `solve_function` przekazujemy do niej funkcję zwracającą *wektor współczynników wielomianu* $p_n$. Do wyświetlenia wykresu używamy biblioteki `matplotlib`. Aby wykres był dokładniejszy, zagęściliśmy zbiór argumentów.

```python
def show_func_and_poly(n, solve_function):
    a_list = solve_function(n)
    def p(x): return calculate_polynomial(a_list, x)
    
    dense_x_range = np.arange(x_range[0], x_range[1], 0.001)
    func_y = f(dense_x_range)
    poly_y = p(dense_x_range)
    plt.plot(dense_x_range, func_y, 'r')
    plt.plot(dense_x_range, poly_y, 'g')
    plt.show()
```

### Sprawdźmy działanie naszej funkcji dla wartości $n$  podanych w treści zadania.

```python
for n in n_cases:
    show_func_and_poly(n, calculate_factors)
```

Widzimy, że na końcach przedziału wykres wielomianu i funkcji znacząco się od siebie różnią, podczas gdy w środkowych częściach przedziału są prawie identyczne (zwłaszcza ze wzrostem $n$). Odpowiada za to tak zwany **efekt Rungego**, czyli zmniejszenie dokładności mimo zwiększania ilości węzłów. Występuje on podczas interpolacji wielomianami i gdy węzły interpolacyjne są równoodległe.

## 1 b)

Chcemy teraz obliczyć różnicę pomiędzy wartościami funkcji i wielomianu oraz przedstawić je na wykresie.

```python
def calc_rest(n):
    a_list = calculate_factors(n)
    def p(x): return calculate_polynomial(a_list, x)
    def rest(x): return f(x) - p(x)
    dense_x_range = np.arange(x_range[0], x_range[1], 0.001)
    rest_result = rest(dense_x_range)
    plt.plot(dense_x_range, rest_result, 'b')
    plt.show()


calc_rest(29)
```

Potwierdzają się nasze przypuszczenia:
- różnice na krańcach przedziału są ogromne i symetryczne.
- różnice w środkowych częściach przedziału są prawie równe 0.

Widzimy więc, że zaszedł tutaj *Efekt Rungego*.

# Ex. 2)

Dla danej funkcji i przedziału z poprzedniego zadania oraz $n = 15$ znajdź wielomian interpolacyjny w węzłach Czebyszewa, danych wzorem $x_k = \frac{1}{2}(a+b) + \frac{1}{2}(b-a)cos(\frac{2k-1}{2n}\pi)$
 gdzie $k = 1, .., n$, a $a, b$ są odpowiednio początkiem i końcem przedziału interpolacji.

Narysuj wykres wynikowego wielomianu. W jaki sposób zmienił się błąd interpolacji w
porównaniu do wyników z poprzedniego zadania? Czemu tak się stało?

## Czebyszew function

```python
def czeb(a, b, k, n):
    return 0.5 * (a + b) + 0.5 * (b - a) * np.cos(((2*k-1)/(2*n))*np.pi)
```

## Solution

Tym razem zmienimy podejście tworzenia węzłów. W dalszym ciągu stworzymy ich $n+1$ jednak tym razem nie będą rozłożone równomiernie. Ich rozkład będzie determinować funkcja czebyszewa, której wzór mamy podany powyżej. Do obliczenia współczynników wielomianu $p_n$ i zaprezentowania wyników na wykresie użyjemy funkcji `show_func_and_poly` zdefiniowanej poniżej (podmienimy jedynie metodę generowania współrzędnych x węzłów).

```python
def calculate_czeb_factors(n):
    a = x_range[0]
    b = x_range[1]
    x_list = np.array([czeb(a, b, k, n) for k in range(1, n+1)])
    y_list = f(x_list)
    
    vander_matrix = np.vander(x_list, increasing=True)
    a_list = np.linalg.solve(vander_matrix, np.transpose(y_list))
    return a_list
    
       
show_func_and_poly(15, calculate_czeb_factors)
```

Widzimy od razu, że ogromne odchylenia na krańcach przedziału zniknęły. Widać, że wielomian (kolor <font color='green'>zielony</font>) również jest bardzo zbliżony do funkcji f (kolor <font color='red'>czerwony</font>). Dzieje się tak ponieważ:
- węzły nie są rozłożone równomiernie.
- węzły zagęszczają się ku krańcom przedziału - zostały specjalnie dostosowane do interpolacji wielomianowej.

Powyższe kroki stosuje się aby uniknąć *efektu Rungego*. W powyższym przykładzie widać ich skuteczność.

# Ex. 3)

Mając daną elipsę w postaci parametrycznej

$\begin{cases}
x = a*cos(t),\\
y = b*sin(t)
\end{cases}$

gdzie $t \in [0, 2\pi]$ wykonaj jej interpolację sześciennymi funkcjami sklejanymi dla 10 równoodległych wartości $t_i$. Interpolację wykonaj osobno dla zbiorów punktów $(t_i, x_i)$ oraz $(t_i, y_i)$, a następnie narysuj przybliżenie powyższej elipsy używając wyznaczonych funkcji dla 30 punktów.

## define functions and $t$ range

```python
def x(a, t): return a*np.cos(t)
def y(b, t): return b*np.sin(t)

t = [0, 2*np.pi]
```

Bez straty ogólności przyjmiemy $ a = 1.5 $ oraz $ b = 3.0 $, w rzeczywistości mogą to być dowolne stałe.

```python
a = 1.5
b = 3.0
```

Za pomocą sparametryzowanego układu równań, obliczymy wartości $x_i$ oraz $y_i$ dla $n$ równomiernie rozłożonych wartości t z przedziału $[0, 2\pi]$.

```python
def generate_sets(a, b, n):
    t_set = np.linspace(t[0], t[1], n)
    x_set = x(a, t_set)
    y_set = y(b, t_set)
    return t_set, x_set, y_set
```

Użyjemy pakietu `interpolate` z biblioteki `scipy`, aby wykonać interpolację sześciennymi funkcjami sklejanymi. Konkretnie w tym celu użyjemy klasy `CubicSpline`. 

```python
def cube_interpolate(args_set, values_set, inter_args):
    cs = CubicSpline(args_set, values_set, bc_type='periodic')
```

Najpierw wygenerujemy zbiór wartości $t, x, y$ dla $n = 10$ oraz $n = 30$ (Do narysowania dokładniejszego wykresu)

```python
t_set, x_set, y_set = generate_sets(a, b, 10)
t_plot_set, x_plot_set, y_plot_set = generate_sets(a, b, 30)
```

Teraz znajdziemy **funkcję interpolacyjną**, która jako argument przyjmuje $t$ i zwraca odpowiednio $x$ lub $y$ (w zależności jakiego zbioru użyliśmy do jej stworzenia). W tym celu posłuży nam wyżej wspomniana klasa `CubicSpline`, która realizuje model sześciennych funkcji sklejanych.

Najpierw zrobimy to dla zbioru punktów $(t_i, x_i)$ a następnie dla $(t_i, y_i)$.

```python
cs_generated_by_x = CubicSpline(t_set, x_set, bc_type='periodic')
cs_generated_by_y = CubicSpline(t_set, y_set, bc_type='periodic')
```

Mając powyższe funkcje możemy przedstawić nasz rezultat na wykresie. Zrobimy to dla $n = 30$ (skorzystamy z wyżej wygenerowanego zbioru `t_plot_set`)

```python
plt.plot(cs_generated_by_x(t_plot_set), cs_generated_by_y(t_plot_set), 'r')
plt.show()
```

Od razu widzimy, że wygenerowana krzywa (nie funkcja!) jest bardzo podobna do tej z treści zadania. Dla porównania pokażemy dokładną elipsę na wykresie (kolor <font color='blue'>niebieski</font>), a później obie krzywe jednocześnie (ta z interpolacji będzie zaznaczona kolorem <font color='red'>czerwonym</font>)

```python
plt.plot(x_plot_set, y_plot_set, 'b')
plt.show()
```

```python
plt.plot(x_plot_set, y_plot_set, 'b')
plt.plot(cs_generated_by_x(t_plot_set), cs_generated_by_y(t_plot_set), 'r')
plt.show()
```

Kolor niebieski jest ledwo dostrzegalny. Dzieje się tak dlatego, że krzywa pochodząca z interpolacji jest **bardzo** dobrym przybliżeniem rzeczywistości i w wielu miejscach krzywe po prostu się pokrywają.

# Wnioski:

    - Metoda interpolacji wielomianami jest bardzo przydatna, jednak ma pewne wady (Efekt Rungego).
    - Istnieją pewne metody radzenia sobie z powyższymi wadami, np. rozłożenie węzłów zgodnie z miejscami
    zerowymi wielomianu Czebyszewa 
    - Metoda interpolacji sześciennymi funkcjami sklejanymi pozwala nam interpolować nie tylko funkcje ale też krzywe, jest także bardzo dokładna
