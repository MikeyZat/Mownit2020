# Ex. 1)


Dla funkcji  $ f(x) = \frac{1}{1 + x^2} $  i $ n = 5, 10, 15 $ wylicz wielomian interpolacyjny $ p_n$ z $ n + 1 $ węzłami równoodległymi w przedziale $ [−5, 5] $.

    a) Narysuj funkcje $f(x)$ oraz $p_n(x)$
    b) Wylicz i narysuj reszty interpolacji $ e(x) = f(x) − p_n(x)$ w 30 równoodległych
    punktach powyższego przedziału.

## Define function and x range

Zapisujemy przedział dla x-ów, wartości n oraz funkcję z treści zadania 

```python
def f(x): return 1/(1 + x**2)

n_cases = [5, 10, 15]
x_range = [-5, 5]
```

## 1 a)

- Mając przedział $ [-5, 5] $, dzielimy go na $ n $ równych części, tworząc $ n + 1 $ węzłów (linie 1-2 funkcji calculate_factors). Następnie dla każdego węzła o współrzędnej x wyliczamy wartość f(x) (używając funkcji zdefiniowanej powyżej).
- Przekazując wektor współrzędnych x węzłów do funkcji `vander` z biblioteki `numpy`, otrzymujemy jako wynik macierz Vandermonde'a, której użyjemy do obliczenia współczynników naszego wielomianu. 
- Przekazujemy macierz Vandermonde'a (`vander_matrix`) wraz z wartościami y węzłów (`y_list` jako wektor) do funkcji `numpy.linalg.solve` i jako wynik otrzymujemy **wektor** zawierający **współczynniki wielomianu interpolacyjnego $ p_n $**

```python
def calculate_factors(n):
    x_list = np.arange(x_range[0], x_range[1], (x_range[1] - x_range[0]) / n)
    x_list = np.append(x_list, x_range[1])
    y_list = f(x_list)
    
    vander_matrix = np.vander(x_list, increasing=True)
    a_list = np.linalg.solve(vander_matrix, np.transpose(y_list))
    return a_list
```

Mając wektor współczynników, funkcja poniżej zwraca wartość wielomianu w punkcie x

```python
def calculate_polynomial(a_list, x):
    sum = 0
    for i in range(len(a_list)):
        sum += a_list[i]*(x**i)
    return sum
```

Teraz mając wszystkie potrzebne dane, chcemy na wykresie przedstawić zarówno funkcję $ f(x) $ jak i wielomian interpolacyjny $p_n$. Poniższa funkcja realizuje tę zadanie. Jako parametr `solve_function` przekazujemy do niej funkcję zwracającą wektor współczynników wielomianu $p_n$. Do wyświetlenia wykresu używamy biblioteki `matplotlib`. Aby wykres był dokładniejszy, zagęściliśmy zbiór argumentów.

```python
def show_func_and_poly(n, solve_function):
    a_list = solve_function(n)
    def p(x): return calculate_polynomial(a_list, x)
    
    dense_x_range = np.arange(x_range[0], x_range[1], 0.001)
    func_y = f(dense_x_range)
    poly_y = p(dense_x_range)
    plt.plot(dense_x_range, func_y, 'r')
    plt.plot(dense_x_range, poly_y, 'g')
    plt.show()
```

### Sprawdźmy działanie naszej funkcji dla wartości $n$  podanych w treści zadania.

```python
    
for n in n_cases:
    show_func_and_poly(n, calculate_factors)
   
```

Widzimy, że na końcach przedziału wykres wielomianu i funkcji znacząco się od siebie różnią, podczas gdy w środkowych częściach przedziału są prawie identyczne. Dzieje się tak, ponieważ ... TODO

## 1 b)

Chcemy teraz obliczyć różnicę pomiędzy wartościami funkcji i wielomianu oraz przedstawić je na wykresie.

```python
def calc_rest(n):
    a_list = calculate_factors(n)
    def p(x): return calculate_polynomial(a_list, x)
    def rest(x): return f(x) - p(x)
    dense_x_range = np.arange(x_range[0], x_range[1], 0.001)
    rest_result = rest(dense_x_range)
    plt.plot(dense_x_range, rest_result, 'b')
    plt.show()


calc_rest(29)
```

Potwierdzają się nasze przypuszczenia:
- różnice na krańcach przedziału są ogromne i symetryczne
- różnice w środkowych częściach przedziału są prawie równe 0

# Ex. 2)

Dla danej funkcji i przedziału z poprzedniego zadania oraz $n = 15$ znajdź wielomian interpolacyjny w węzłach Czebyszewa, danych wzorem $x_k = \frac{1}{2}(a+b) + \frac{1}{2}(b-a)cos(\frac{2k-1}{2n}\pi)$
 gdzie $k = 1, .., n$, a $a, b$ są odpowiednio początkiem i końcem przedziału interpolacji.

Narysuj wykres wynikowego wielomianu. W jaki sposób zmienił się błąd interpolacji w
porównaniu do wyników z poprzedniego zadania? Czemu tak się stało?

## Czebyszew function

```python

def czeb(a, b, k, n):
    return 0.5 * (a + b) + 0.5 * (b - a) * np.cos(((2*k-1)/(2*n))*np.pi)
```

## Solution

Tym razem zmienimy podejście tworzenia węzłów. W dalszym ciągu stworzymy ich $n+1$ jednak tym razem nie będą rozłożone równomiernie. Ich rozkład będzie determinować funkcja czebyszewa, której wzór mamy podany powyżej. Do obliczenia współczynników wielomianu $p_n$ i zaprezentowania wyników na wykresie użyjemy funkcji `show_func_and_poly` zdefiniowanej poniżej (podmienimy jedynie metodę generowania współrzędnych x węzłów).

```python
def calculate_czeb_factors(n):
    a = x_range[0]
    b = x_range[1]
    x_list = np.array([czeb(a, b, k, n) for k in range(1, n+1)])
    y_list = f(x_list)
    
    vander_matrix = np.vander(x_list, increasing=True)
    a_list = np.linalg.solve(vander_matrix, np.transpose(y_list))
    return a_list
    
       
show_func_and_poly(15, calculate_czeb_factors)
```
